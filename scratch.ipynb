```python
import json
import os
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
directory = '/home/msc24h18/Documents/msc_project/neurosim_configs/test8/'
json_files = [f for f in os.listdir(directory) if f.endswith('.json')]
maxConductance = []
minConductance = []
pulseWidth = []
for f in json_files:
    data = json.load(open(directory + f))
    maxConductance.append(data['device-params']['RealDevice']['maxConductance'])
    minConductance.append(data['device-params']['RealDevice']['minConductance'])
    pulseWidth.append(data['device-params']['RealDevice']['readPulseWidth'])

maxConductance = np.array(maxConductance)
minConductance = np.array(minConductance)
pulseWidth = np.array(pulseWidth)

onOffRatio = maxConductance/minConductance

fig, ax = plt.subplots()
ax.scatter(pulseWidth, maxConductance, label = 'Max Conductance')
ax.scatter(pulseWidth, minConductance, label = 'Min Conductance')
ax.set(xscale='log', xlabel='Pulse Width (s)', ylabel='Max Conductance (S)', ylim=(0, 4e-10), title='ID301XR1000Oct')
ax.legend()
#plt.savefig('presentation-plots/fig14_ID301XR1000Oct.png')
```

```python
fig, ax = plt.subplots()
ax.scatter(pulseWidth, onOffRatio)
ax.set(xscale='log', xlabel='On/Off Ratio', ylabel='Pulse Width (s)', title='ID301XR1000Oct')
```
