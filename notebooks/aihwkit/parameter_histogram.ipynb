```python
import pandas as pd
import matplotlib.pyplot as plt
import msc_project.utils.data_utils as du

plt.style.use('ggplot')
```

```python
df = pd.read_csv('/scratch/msc24h18/msc_project/results/test52/test52.csv')
```

```python
fig, axs = plt.subplots(2, 3, figsize=(20,10), sharey=False)

numlevels_ltp_list = df['num_LTP']
numlevels_ltd_list = df['num_LTD']
nl_ltp_list = df['NL_LTP']
nl_ltd_list = -df['NL_LTD']
sp_list = df['symmetry_point']

gran_list = df['granularity']
gran_diff_list = df['granularity_diff']
w_min_list = -df['w_min']
w_max_list = df['w_max']

axs[0][0].hist(numlevels_ltp_list, bins=10, color="C0", alpha=0.7, label="LTP")
axs[0][0].hist(numlevels_ltd_list, bins=10, color="C1", alpha=0.7, label="LTD")
axs[0][0].legend()
axs[0][0].set(ylabel = "Count", xlabel = "Number of Conductance States", title = "Number of States Distribution")

axs[0][1].hist(nl_ltp_list, bins=10, color="C2", alpha=0.7, label="LTP")
axs[0][1].hist(nl_ltd_list, bins=10, color="C3", alpha=0.7, label="LTD")
axs[0][1].legend()
axs[0][1].set(xlabel = "Nonlinearity Parameter (abs.)", title = "Nonlinearity Distribution (NeuroSim)")

axs[0][2].hist(sp_list, bins=10, color="C4", alpha=0.7)
axs[0][2].set(xlabel = "Symmetry Point", title = "Symmetry Point Distribution")

axs[1][0].hist(gran_list, bins=10, color="C0", alpha=0.7)
axs[1][0].set(ylabel = "Count", xlabel = "Granularity", title = "Granularity Distribution (AIHWKIT)")

axs[1][1].hist(gran_diff_list, bins=10, color="C1", alpha=0.7)
axs[1][1].set(xlabel = "Asymmetry (granularity up - granularity down)", title = "Asymmetry Distribution (AIHWKIT)")

axs[1][2].hist(w_min_list, bins=10, color="C2", alpha=0.7, label="$|w_{min}|$")
axs[1][2].hist(w_max_list, bins=10, color="C3", alpha=0.7, label="$|w_{max}|$")
axs[1][2].set(xlabel = "Weight", title = "$w_{min/max}$ Distribution (AIHWKIT)")
axs[1][2].legend()

plt.tight_layout()
plt.savefig('tmp.png')
plt.show()
```
