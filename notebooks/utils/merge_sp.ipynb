```python
import os
import numpy as np
import pandas as pd
import msc_project.utils.data_utils as du
```

```python
DIR='/scratch/msc24h18/msc_project/results/symmetry_point_1'
```

```python
png_files = du.get_files(DIR, 'png')
sp_data = {'device_id':[], 'test_date':[], 'test_time':[], 'symmetry_point':[]}
for f in png_files:
    title= os.path.basename(os.path.dirname(f))
    device_id = title.split('_')[2].split('(')[0]
    dt = title.split('_')[2].split(']')[-1]
    date = dt[:8]
    time = dt[8:]
    symmetry_point = float(os.path.basename(f).split('_')[-1][:-4])

    date = date[-2:] + '.' + date[4:6] + '.' + date[:4]
    time = time[:2] + ':' + time[2:4] + ':' + time[4:]

    if time == "11:05:16":
        time = "11:05:17"

    sp_data['device_id'].append(device_id)
    sp_data['test_date'].append(date)
    sp_data['test_time'].append(time)
    sp_data['symmetry_point'].append(symmetry_point)
    print(device_id, date, time, symmetry_point)
```

```python
sp_data = pd.DataFrame(sp_data)
data = pd.read_csv('/scratch/msc24h18/msc_project/results/test25/test25.csv')
```

```python
merged = pd.merge(data, sp_data, on=['device_id', 'test_date', 'test_time'], how='left')
merged
```
