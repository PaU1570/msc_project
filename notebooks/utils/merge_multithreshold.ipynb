```python
import pandas as pd
import numpy as np
```

```python
DATA_PATH = '/scratch/msc24h18/msc_project/results/test48/'
df1 = pd.read_csv(DATA_PATH + 'symmetric_0_noise_1.csv')
df2 = pd.read_csv(DATA_PATH + 'asymmetric_0_noise_1.csv')
df1.head()
```

```python
columns={'test_acc': 'test_acc_sym',
    'val_acc': 'val_acc_sym',
    'val_loss': 'val_loss_sym',
    'train_loss': 'train_loss_sym',
    'pulses': 'pulses_sym',
    'pulses_pos': 'pulses_pos_sym',
    'pulses_neg': 'pulses_neg_sym'
}

df1.rename(columns = columns, inplace = True)

for key in columns.keys():
    df1[key + '_asym'] = df2[key]

df1['granularity_diff'] = np.abs(df1['granularity_up'] - df1['granularity_down'])
df1['val_acc_diff'] = df1['val_acc_asym'] - df1['val_acc_sym']
```

```python
df1.to_csv(DATA_PATH + 'combined_0_noise_1.csv', index=False)
```
